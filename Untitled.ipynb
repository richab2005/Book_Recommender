```python
#merge two datasets
merged=ratings.merge(books,on='ISBN')
merged
```

#popularity based recommendation system

```python
num_rating=merged.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'Number of Rates'},inplace=True)
num_rating
```

```python
avg_rating = (merged.groupby('Book-Title')['Book-Rating'] .mean().reset_index())
avg_rating.rename(columns={'Book-Rating':'Average-Rating'},inplace=True)
avg_rating
```

```python
popular_df=num_rating.merge(avg_rating,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Number of Rates'] >= 250].sort_values('Average-Rating',ascending=False).head(50)
popular_df.head()
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Average-Rating','Number of Rates']]
popular_df['Average-Rating'] = popular_df['Average-Rating'].round(1)

popular_df
```

#collaborative Filtering Based Recommendation

```python
x=merged.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_user=x[x].index
padhe_likhe_user #these are the users who rated above 200 books
```

```python
filtered_rating=merged[merged['User-ID'].isin(padhe_likhe_user)]
filtered_rating.head()
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>50 #ashe book jyana 50 peksha jast vote bhetle
famous_books=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt=pt.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
#we calculate euclide distance of each row with remaining all rows
```

```python
similarity_score.shape
```

```python
def recomend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1] ,reverse=True)[1:8]

    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)
    return data

```

```python
recomend('Message in a Bottle')
```

```python
recomend('The Notebook')
```

```python
recomend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
